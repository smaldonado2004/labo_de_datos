### La magia de Numpy.
La biblioteca Numpy reproduce muchas funcionalidades de Matlab (un lugar donde se hacen operaciones matematicas).
La mayoría de las operaciones con vectores de Numpy se hacen **coordenada a coordenada**. Esto permite en muchos casos **evitar usar ciclos o ciclos anidados** y realizarlos con un solo comando.

La diferencia de tiempo de ejecución entre utilizar operaciones apropiadas de Numpy o hacerlo mediante ciclos puede ser muy grande.

Ejecutar los siguientes comandos e interpretar los resultados.

```python
v = np.array([1,2,np.e,7]) #np.e es el nro. de euler 2,71
w = np.array([1.2, np.pi, 4, 5])

print("v = ", v) #ojo como se conecta info del res con res con coma
print("w = ", w)
print("v + w = ", v + w)
```

```python
#todas las operaciones son componente a componente:
print ("v**2 = ", v**2)
print ("v%2 = ", v%2)
```

```python
print ("v > 3?", v > 3)
print ("w < 3.5?", w < 3.5)
```

Algunas funciones de numpy pueden usarse también como métodos

Las operaciones lógicas **or y and no se pueden aplicar a vectores**. Debemos usar los símbolos **|

```python
print((v > 3) or (w < 3.5))
print((v > 3) and (w < 3.5))
#TIRA ERROR PQ NO SE PUEDE USAR OR Y AND SE USA | &
```

El **producto usual de matrices se realiza con el comando @** .

Vamos a multiplicar estas dos matrices sencillas:

$$A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}, \quad B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}$$

El resultado será una nueva matriz $C$ de $2 \times 2$. Calculamos cada posición:

*   **Fila 1 de A × Columna 1 de B (Posición $C_{11}$):**
$$(1 \times 5) + (2 \times 7) = 5 + 14 = 19$$

*   **Fila 1 de A × Columna 2 de B (Posición $C_{12}$):**
$$(1 \times 6) + (2 \times 8) = 6 + 16 = 22$$

*   **Fila 2 de A × Columna 1 de B (Posición $C_{21}$):**
$$(3 \times 5) + (4 \times 7) = 15 + 28 = 43$$

*   **Fila 2 de A × Columna 2 de B (Posición $C_{22}$):**
$$(3 \times 6) + (4 \times 8) = 18 + 32 = 50$$

**Resultado final:**
$$A @ B = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}$$


**A.T** es la transpuesta de A. Es dar vuela la matriz:
A =
[[1, 2, 3],
 [4, 5, 6]]

A.T =
[[1, 4],
 [2, 5],
 [3, 6]]

**Ejercicio.** Definir una función que calcule la norma-2 de un vector y verificarla en los vectores $v_0$ y $v_1$ del ejercicio anterior.

**La norma es la raiz de la suma de cada componente al cuadrado**

```python
def norma(vector):
  elevar = vector**2
  suma = elevar.sum() #lo que sumo es lo que elevo
  res = np.sqrt(suma)

  return res

print("normaV0", norma(v0))
print("normaV1", norma(v1))
```

1. Generar 1000 puntos al azar en el cuadrado de vértices $(0,0), (0,1), (1,0), (1,1)$ y calcular que proporción esta a distancia menor de 1 del origen $(0,0)$.
2. Utilizar el resultado anterior para dar una aproximación de $\pi$.


```python
#EJERCICIO 1
#Quiero que mis puntos cumplan: 0 ≤ x ≤ 1 , 0 ≤ y ≤ 1
#Genero 1000 valores al azar del 0 al 1
valores = np.linspace(0,1,1000)

#Genero puntos al azar a partir de valores:
x = np.random.choice(valores, size=1000)
y = np.random.choice(valores, size=1000)

#Busco los puntos que estan a distancia menor a 1 del origen con norma:
distancia = (np. sqrt(x**2+y**2)) < 1

#Hallo la proporcion (promedio de dist. menor a uno
proporcion = distancia.sum() / 1000
print(proporcion)
```

La condicion x^2 + y^2 < 1 es un circulo de radio uno, el area completa del circulo de radio uno es pi.
Como solo estoy viendo x ≥ 0 e y ≥ 0 que seria un cuarto del circulo.

Entonces, **proporcion = pi / 4**, despejando pi que es lo que queremos aproximar (dar un valor parecido a 3.14), nos queda: pi = 4 * proporcion

```python
pi_aprox = 4 * proporcion
print(pi_aprox)
```
