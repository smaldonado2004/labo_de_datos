8. La magia de Numpy La biblioteca Numpy reproduce muchas funcionalidades de Matlab. La
mayor´ıa de las operaciones con vectores de Numpy se hacen coordenada a coordenada. Esto
permite en muchos casos evitar usar ciclos o ciclos anidados y realizarlos con un solo comando.
Ejecutar los siguientes comandos e interpretar los resultados.

```python
np . set_printoptions ( precision =2 , suppress = True )
#La anterior linea no es necesaria para q el codigo funcine, solo se usa para mejorar como se muestran los nros.
  #Precision = 2: muestra solo 2 decimales
  #supress = True: evita notacion cientifica (por ejemplo 1e-05)
v = np . array ([1 ,2 , np .e ,7])
w = np . array ([1.2 , np . pi , 4 , 5])
print ("v+w = ", v + w )
print ("v*w = ", v * w )
```

Podemos probar varios comandos en un ciclo

```python
#Para cada expresion en la lista:
for expr in ["v", "w", "v + w", "v ** 2", "v % 2", "np.sum(v)", "np. sqrt (w)", "v > 3", "w < 3.5 "
]:
  print ( f"{ expr :11s} == { eval ( expr )}")
  #{expr:11s}: expr: la expresion, 11: ocupa 11 espacios, s: string (sirve para alinear la salida a lo q esta antes de ==)
  #eval ( expr ): evalua cada expresion convirtiendola de text. a un res de una op real
```

25. Reproducir estos usos de la funcion np.random.choice() e interpretar que hace.

```python
x = np . array ([" cara ", " ceca "])
s = np . random . choice (x , size =10)
print ("s", s ) #size es desde el uno

y = np . array ([1 ,2 ,3 ,4 ,5 ,6])
t = np . random . choice (y , size =10)
print ("t", t )

u = np . random . choice (y , size =4 , replace = False )
print ("u", u ) #sin repetidos
```

¿Para que sirve el parametro replace=False? Realicen otras pruebas para verificarlo. Comparen con random.choices y random.sample de la biblioteca estandar de Python.

```python
x = np.array([1, 2, 3])

#RANDOM CHOICE SIN REPETIDOS
np.random.choice(x, size=3, replace=False)
```

```python
import random #para usar random choices y sample
#RANDOM SAMPLE: no adimite repetidos
random.sample(list(x), k=3)
```
